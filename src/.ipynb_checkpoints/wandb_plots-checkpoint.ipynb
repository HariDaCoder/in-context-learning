```python
from collections import OrderedDict
import re
import os

import matplotlib.pyplot as plt
import torch
import seaborn as sns
from tqdm.notebook import tqdm
import pandas as pd
import numpy as np

%matplotlib inline
sns.set_theme('notebook', 'darkgrid')
palette = sns.color_palette('colorblind')
wandb_dir = 'wandb_plot_data'
```

```python
for dim in [10, 20, 30, 40, 50]:
    pd_loss = pd.read_csv(os.path.join(wandb_dir, str(dim)+".csv"))
    pd_n_points = pd.read_csv(os.path.join(wandb_dir, str(dim)+"_n_points.csv"))

    steps = pd_loss['Step'].to_numpy()
    plt.figure(figsize=(4,3))
    plt.axhline(sum(range(dim+1)) / dim**2 / 2, ls="--", color="grey")
    plt.axvline(2000 * (dim-5), ls="--", color="grey")
    for trial in ["", "_trial2", "_trial3"]:
        loss_flat = pd_loss[f"linear_regression_{dim}_8head_flat{trial} - overall_loss"].to_numpy()
        loss_curriculum = pd_loss[f"linear_regression_{dim}_8head{trial} - overall_loss"].to_numpy()
        loss_flat_smooth = []
        loss_curriculum_smooth = []
        for i in range(len(loss_flat)):
            window = min(i, 10)
            loss_flat_smooth.append(loss_flat[i-window:i+1].mean())
            loss_curriculum_smooth.append(loss_curriculum[i-window:i+1].mean())
        loss_flat_smooth = np.array(loss_flat_smooth)
        loss_curriculum_smooth = np.array(loss_curriculum_smooth)

        n_points = pd_n_points[f"linear_regression_{dim}_8head - n_points"]
        plt.plot(steps, loss_curriculum_smooth/dim, label='with curriculum', color=palette[0], lw=2)
        plt.plot(steps, loss_flat_smooth/dim, label='without curriculum', color=palette[1], lw=2)
    plt.xlabel('training steps')
    plt.ylabel('loss')
    plt.ylim(-.1, 1.1)
    plt.xticks(
        [0, 100000, 200000, 300000, 400000, 500000],
        ["0", "100k", "200k" , "300k", "400k" , "500k"]
    )
    if dim == 10:
        plt.xlim(0, 100000)

    #plt.legend()
    plt.savefig(f"figures/curriculum_{dim}.pdf", bbox_inches='tight')
    plt.show()
```

```python
plt.figure(figsize=(0,0))
plt.plot([1], label = 'with curriculum', lw=3, color=palette[0])
plt.plot([1], label = 'without curriculum', lw=3, color=palette[1])
plt.legend()
plt.axis("off")
plt.savefig(f"figures/curriculum_legend.pdf", bbox_inches='tight')
plt.show()
```
