```python
#Training points also in the 1d subspace
#We should mostly ignore this plot for the paper. There is no clear pattern here.

x = sampler.sample_xs(n_points = n_points, b_size = 1).squeeze(0)
task = task_sampler()
dir1 = torch.randn(n_dims)
dir1 = dir1/torch.linalg.norm(dir1)
x = math.sqrt(n_dims)*torch.randn(n_points, 1) @ dir1.view(1,-1)
ind = 10

x_test_grid, pred_test_grid, y_test_grid, marks_grid  = gen_vis_data(model.cuda(), x, ind = ind, task = task, dir1 = dir1, dir1_lim = 3*math.sqrt(n_dims), grid_size = 100, batch_size = 64)

x_projs = (x @ dir1.view(-1,1)).view(-1)
y_projs = task.evaluate(x.unsqueeze(0))[0]

plt.plot(marks_grid, y_test_grid, '-o', label = 'ground truth', zorder = 0)
plt.plot(marks_grid, pred_test_grid, '-o', label = 'prediction', zorder = 5)
plt.scatter(x_projs[:ind], y_projs[:ind], marker = 'x', color = 'yellow', zorder=10)
plt.xlabel('distance from origin')
plt.ylabel('function value')
plt.title('Prediction with training points in the same 1-d subspace as the test point')
plt.legend()
plt.show()
```

```python
#plot settings
alpha_ci = 0.5
```

```python
task_sampler = get_task_sampler(task_name, n_dims, batch_size = 1)
num_runs = 40
b_size = 64

model.cuda()
grad_metrics = eval_grads(model, task_sampler, sampler, n_points, num_runs, b_size)

inner_prod_grad_true_agg = aggregate_metrics(grad_metrics[0])
inner_prod_grad_grad_agg = aggregate_metrics(grad_metrics[1])

#metrics is a tuple
#metrics[0] is a tensor of size num_runs X n_points containing average inner products between grad and true w
#metrics[1] is a tensor of size num_runs X n_points containing average inner products between grad and grad

plt.plot(inner_prod_grad_true_agg['mean'], label = 'b/w grad and true w')
plt.fill_between(torch.arange(n_points),inner_prod_grad_true_agg['bootstrap_low'],inner_prod_grad_true_agg['bootstrap_high'], alpha = alpha_ci)

plt.plot(inner_prod_grad_grad_agg['mean'], label = 'b/w grad and true w')
plt.fill_between(torch.arange(n_points),inner_prod_grad_grad_agg['bootstrap_low'],inner_prod_grad_grad_agg['bootstrap_high'], alpha = alpha_ci)

plt.xlabel('number of training points')
plt.ylabel('average inner product')
plt.legend()
plt.show()
```

```python
model.cuda()
num_runs = 40
batch_size = 64

task_sampler = get_task_sampler(task_name, n_dims, batch_size = batch_size)
scales = [1/10.0, 1/8.0, 1/6.0, 1/4.0, 1/2.0,  1, 2, 3, 4, 5, 6, 8, 10]
inds = [n_dims//2, n_dims, 2*n_dims]

scale_metrics = eval_scaled_x_test(model, task_sampler, x_sampler = sampler, inds = inds, scales = scales, num_runs = num_runs, batch_size = batch_size)

#metrics is a dictionary of the form {inds[0]:{scales[0]:metric, scales[1]:metric}, inds[1]: ...} 
#where metric is a tensor of size num_runs*batch_size.

scales = []
normalized_losses = []
higher_cis = []
lower_cis = []
ind = 2*n_dims
for scale in sorted(scale_metrics[ind].keys()):
    scales.append(scale)
    metrics_agg = aggregate_metrics(scale_metrics[ind][scale].view(-1,1))
    
    normalized_losses.append(metrics_agg['mean'][0]/((scale**2)*n_dims))
    higher_cis.append(metrics_agg['bootstrap_high'][0]/((scale**2)*n_dims))
    lower_cis.append(metrics_agg['bootstrap_low'][0]/((scale**2)*n_dims))
    
plt.plot(scales, normalized_losses, 'o-')
plt.fill_between(scales, lower_cis, higher_cis, alpha = alpha_ci)
plt.xlabel('scale')
plt.ylabel('normalized squared loss')
plt.show()
```
